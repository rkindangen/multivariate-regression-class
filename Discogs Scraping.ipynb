```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import time
import random
import numpy as np
import os
```

```python
album_release_year = []
album_name = []
album_link = []
album_artist_link = []
album_country = []
```

```python
for i in range(1, 10):
    url = "https://www.discogs.com/search/?sort=have%2Cdesc&ev=em_rs&format_exact=Vinyl&layout=sm&page=" + str(i)
    time.sleep(10+2*random.random())
    response = requests.get(url, headers = {'User-agent': 'Mozilla/5.0'}, verify = False)
    page = response.text
    soup = BeautifulSoup(page, "lxml")

    albums =  soup.find(id = 'search_results').find_all(class_= "card_body")
    for album in albums:
        try: 
            album_release_year.append(album.find_all(class_="card_release_year")[0].text)
        except:
            album_release_year.append(None)
        try:
            album_name.append(album.find_all("a", class_="search_result_title")[0].text)
        except:
            album_name.append(None)
        try: 
            album_link.append('https://www.discogs.com' + album.find_all("a", class_="search_result_title")[0].get("href"))
        except:
            album_link.append(None)
        try:
            album_artist_link.append('https://www.discogs.com' + album.find_all("a")[0].get("href"))
        except:
            album_artist_link.append(None)
        try: 
            album_country.append(album.find_all(class_ = "card_release_country")[0].text)
        except:
            album_country.append(None)
    
```

```python
albums = pd.DataFrame({'album_release_year':album_release_year,'album_name':album_name, 'album_link':album_link, 'album_artist_link':album_artist_link, 'album_artist_link':album_artist_link})

```

```python
albums.info()
```

Scraping total number of albums per artist

```python
total_albums_artist = []
```

```python
for artist in albums['album_artist_link']:
    url = artist
    response = requests.get(url, headers = {'User-agent': 'Mozilla/5.0'}, verify = False)
    time.sleep(2+1*random.random())
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    
    try:
        total_albums_artist.append(int(soup.find(id = "discography_wrapper").find_all("span", class_ = "facet_count")[1].text))
    except:
        total_albums_artist.append(None)
        
```

```python
albums['total_albums_artist'] = total_albums_artist
```

Scraping first release link for each album

```python
first_album_link = []
```

```python
for album in albums['album_link']:
    url = album
    response = requests.get(url, headers = {'User-agent': 'Mozilla/5.0'}, verify = False)
    time.sleep(5+1*random.random())
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    
    try:
        first_album_link.append('https://www.discogs.com' + soup.find('section', id='versions').find('table').find_all('a')[1].get("href"))
    except:
        first_album_link.append(None)
```

```python
users_have = []
users_want = []
user_rating = []
total_rating = []
lowest_price = []
median_price = []
highest_price = []
total_versions = []
```

```python
for link in first_album_link: 
    url = link
    response = requests.get(url, headers = {'User-agent': 'Mozilla/5.0'}, verify = False)
    time.sleep(2+1*random.random())
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    
    try: 
        users_have.append(int(re.search("[0-9]+", soup.find("section", id = "release-stats").find_all('li')[0].text).group(0)))
    except: 
        users_have.append(None)
    try: 
        users_want.append(int(re.search("[0-9]+", soup.find("section", id = "release-stats").find_all('li')[1].text).group(0)))
    except: 
        users_want.append(None) 
    try: 
        user_rating.append(float(re.search("[0-9]+\.?[0-9]+", soup.find("section", id = "release-stats").find_all('li')[2].text).group(0)))
    except: 
        user_rating.append(None) 
    try:
        total_rating.append(int(re.search("[0-9]+", soup.find("section", id = "release-stats").find_all('li')[3].text).group(0)))
    except:
        total_rating.append(None)
    try:
        lowest_price.append(float(re.search("[0-9]+\.?[0-9]+", soup.find("section", id = "release-stats").find_all('li')[5].text).group(0)))
    except:
        lowest_price.append(None)
    try:
        median_price.append(float(re.search("[0-9]+\.?[0-9]+", soup.find("section", id = "release-stats").find_all('li')[6].text).group(0)))
    except:
        median_price.append(None)
    try:
        highest_price.append(float(re.search("[0-9]+\.?[0-9]+", soup.find("section", id = "release-stats").find_all('li')[7].text).group(0)))
    except:
        highest_price.append(None)
    try:
        total_versions.append(int(re.search('[0-9]+', soup.find('section', id = "release-marketplace").find_all('a')[0].text).group(0)))
    except:
        total_versions.append(None)
        
    
                      
                      
```

```python
users_price_df = pd.DataFrame({'users_have': users_have, 'users_want': users_want, 'user_rating': user_rating, 'total_rating': total_rating, 'lowest_price': lowest_price, 'median_price': median_price, 'highest_price': highest_price, 'total_versions': total_versions })
```

```python
users_price_df.head()
```

```python
#concatenate all dataframes
df_final = albums.join(users_price_df)
```

```python
df_final.head()
```

```python
#calculate years since release 
df_final['years_since_release'] = df_final['album_release_year'].apply(lambda x: 2022 - int(x))
```

```python
df_final.to_csv('df_final.csv')
```
